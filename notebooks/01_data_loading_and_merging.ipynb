## Notebook 01 — Data Loading, Merging & Initial Exploration

---

This notebook covers:
1. Loading all four raw CSV datasets
2. Printing shapes of each dataset
3. Merging train (transaction + identity) via left join on `TransactionID`
4. Merging test (transaction + identity) via left join on `TransactionID`
5. Printing merged shapes
6. Previewing first 5 rows of merged train
7. Analyzing fraud label distribution

> **Dataset Source:** IEEE-CIS Fraud Detection (Kaggle)

---
## 📦 Step 0 — Import Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('Libraries loaded successfully ✅')
```

---
## 📂 Step 1 — Load All Four Datasets

```python
DATA_PATH = '../Datasets/'

print('Loading datasets... (this may take a moment due to large file sizes)')

train_transaction = pd.read_csv(DATA_PATH + 'train_transaction.csv')
print('✅ train_transaction.csv loaded')

train_identity = pd.read_csv(DATA_PATH + 'train_identity.csv')
print('✅ train_identity.csv loaded')

test_transaction = pd.read_csv(DATA_PATH + 'test_transaction.csv')
print('✅ test_transaction.csv loaded')

test_identity = pd.read_csv(DATA_PATH + 'test_identity.csv')
print('✅ test_identity.csv loaded')
```

---
## 📐 Step 2 — Print Shapes of All Four Datasets

```python
print('=' * 50)
print('         DATASET SHAPES (rows x columns)        ')
print('=' * 50)
print(f'  train_transaction : {train_transaction.shape}')
print(f'  train_identity    : {train_identity.shape}')
print(f'  test_transaction  : {test_transaction.shape}')
print(f'  test_identity     : {test_identity.shape}')
print('=' * 50)
```

---
## 🔗 Step 3 — Merge Train Datasets (Left Join on `TransactionID`)

```python
train = pd.merge(
    train_transaction,
    train_identity,
    on='TransactionID',
    how='left'
)

print('✅ Train merge complete (train_transaction LEFT JOIN train_identity)')
```

---
## 🔗 Step 4 — Merge Test Datasets (Left Join on `TransactionID`)

```python
test = pd.merge(
    test_transaction,
    test_identity,
    on='TransactionID',
    how='left'
)

print('✅ Test merge complete (test_transaction LEFT JOIN test_identity)')
```

---
## 📐 Step 5 — Print Merged Dataset Shapes

```python
print('=' * 50)
print('       MERGED DATASET SHAPES (rows x columns)   ')
print('=' * 50)
print(f'  Merged Train : {train.shape}')
print(f'  Merged Test  : {test.shape}')
print('=' * 50)
```

---
## 👀 Step 6 — Preview First 5 Rows of Merged Train

```python
print('First 5 rows of merged train dataset:\n')
train.head()
```

---
## 🎯 Step 7 — Fraud Distribution (Value Counts + Percentage)

```python
fraud_counts = train['isFraud'].value_counts()
fraud_pct    = train['isFraud'].value_counts(normalize=True) * 100

fraud_summary = pd.DataFrame({
    'Label'      : ['Not Fraud (0)', 'Fraud (1)'],
    'Count'      : [fraud_counts[0], fraud_counts[1]],
    'Percentage' : [f'{fraud_pct[0]:.2f}%', f'{fraud_pct[1]:.2f}%']
})

print('=' * 45)
print('         FRAUD LABEL DISTRIBUTION              ')
print('=' * 45)
print(fraud_summary.to_string(index=False))
print('=' * 45)
print(f'\nTotal Transactions : {len(train):,}')
print(f'Fraudulent         : {fraud_counts[1]:,}  ({fraud_pct[1]:.2f}%)')
print(f'Legitimate         : {fraud_counts[0]:,}  ({fraud_pct[0]:.2f}%)')
print(f'\n⚠️  Dataset is heavily imbalanced!')
```
