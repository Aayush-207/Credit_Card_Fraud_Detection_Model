## Notebook 03 — Initial Feature Reduction

---

This notebook performs the first round of feature reduction:
1. Load and merge train data
2. Drop all columns with **>80% missing values**
3. Print shape before and after
4. Print number of columns dropped and remaining
5. Print updated memory usage

> **Note:** No imputation is performed at this stage.

---
## 📦 Step 0 — Import Libraries & Load Merged Train Data

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

DATA_PATH = '../Datasets/'

print('Loading and merging datasets...')
train_transaction = pd.read_csv(DATA_PATH + 'train_transaction.csv')
train_identity    = pd.read_csv(DATA_PATH + 'train_identity.csv')
train = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
del train_transaction, train_identity

print(f'✅ Merged train loaded — shape: {train.shape}')
```

---
## 📋 Step 1 — Baseline Stats (Before Reduction)

```python
shape_before  = train.shape
mem_before_mb = train.memory_usage(deep=True).sum() / (1024 ** 2)

# Compute missing % per column
missing_pct = train.isnull().mean() * 100

# Identify columns to drop (>80% missing)
cols_to_drop = missing_pct[missing_pct > 80].index.tolist()

print('=' * 55)
print('        BEFORE FEATURE REDUCTION                  ')
print('=' * 55)
print(f'  Shape              : {shape_before[0]:,} rows x {shape_before[1]} cols')
print(f'  Memory Usage       : {mem_before_mb:.2f} MB')
print(f'  Columns to Drop    : {len(cols_to_drop)}  (those with >80% missing)')
print('=' * 55)
print('\nColumns that will be dropped:')
for i, col in enumerate(sorted(cols_to_drop), 1):
    pct = missing_pct[col]
    print(f'  {i:>3}. {col:<20}  {pct:.2f}% missing')
```

---
## ✂️ Step 2 — Drop Columns with >80% Missing

```python
train_reduced = train.drop(columns=cols_to_drop)

print(f'✅ Dropped {len(cols_to_drop)} columns successfully.')
```

---
## 📊 Step 3 — Results: Shape, Column Count & Memory

```python
shape_after   = train_reduced.shape
mem_after_mb  = train_reduced.memory_usage(deep=True).sum() / (1024 ** 2)
cols_dropped  = shape_before[1] - shape_after[1]
mem_saved_mb  = mem_before_mb - mem_after_mb
mem_saved_pct = (mem_saved_mb / mem_before_mb) * 100

print('=' * 60)
print('           FEATURE REDUCTION SUMMARY                  ')
print('=' * 60)
print(f'  {"":30s} {"Before":>10}   {"After":>10}')
print(f'  {"-"*54}')
print(f'  {"Rows":<30} {shape_before[0]:>10,}   {shape_after[0]:>10,}')
print(f'  {"Total Columns":<30} {shape_before[1]:>10}   {shape_after[1]:>10}')
print(f'  {"Memory (MB)":<30} {mem_before_mb:>10.2f}   {mem_after_mb:>10.2f}')
print(f'  {"-"*54}')
print(f'  {"Columns Dropped":<30} {cols_dropped:>10}')
print(f'  {"Memory Saved (MB)":<30} {mem_saved_mb:>10.2f}   ({mem_saved_pct:.1f}% reduction)')
print('=' * 60)

print(f'\n  ✅ New shape       : ({shape_after[0]:,} rows x {shape_after[1]} columns)')
print(f'  ✅ Columns dropped : {cols_dropped}')
print(f'  ✅ Columns kept    : {shape_after[1]}')
print(f'  ✅ Memory usage    : {mem_after_mb:.2f} MB  (saved {mem_saved_mb:.2f} MB)')
```

---
## 🔍 Step 4 — Verify Remaining Missing Stats

```python
missing_after = train_reduced.isnull().mean() * 100

still_high = (missing_after > 50).sum()
still_any  = (missing_after > 0).sum()
clean      = (missing_after == 0).sum()

print('=' * 55)
print('      REMAINING MISSING DATA OVERVIEW             ')
print('=' * 55)
print(f'  Total remaining columns   : {shape_after[1]}')
print(f'  Columns with no missing   : {clean}')
print(f'  Columns with any missing  : {still_any}')
print(f'  Columns with >50% missing : {still_high}  (need imputation strategy)')
print('=' * 55)
print('  ⚠️  Max missing % in remaining cols: '
      f'{missing_after.max():.2f}%  (col: {missing_after.idxmax()})')
```

---
## 🎯 Step 5 — Fraud Label Distribution

```python
fraud_counts = train_reduced['isFraud'].value_counts()
fraud_pct    = train_reduced['isFraud'].value_counts(normalize=True) * 100

print('=' * 50)
print('       FRAUD LABEL DISTRIBUTION             ')
print('=' * 50)
print(f'  Fraud Count     (1) : {fraud_counts[1]:>10,}')
print(f'  Non-Fraud Count (0) : {fraud_counts[0]:>10,}')
print(f'  Total               : {len(train_reduced):>10,}')
print('-' * 50)
print(f'  Fraud Percentage    : {fraud_pct[1]:>9.2f}%')
print(f'  Non-Fraud Percentage: {fraud_pct[0]:>9.2f}%')
print('=' * 50)
```
