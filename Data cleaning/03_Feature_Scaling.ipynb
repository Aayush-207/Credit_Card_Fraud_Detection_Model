# Feature Scaling & Normalization
## Credit Card Fraud Detection Dataset

This notebook applies scaling/normalization techniques to the features.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load cleaned datasets
card_trans_df = pd.read_csv('../Datasets/card_transdata.csv')
creditcard_df = pd.read_csv('../Datasets/creditcard.csv')

print("Loaded datasets successfully")
```

## 1. Data Before Scaling

```python
# Check data statistics before scaling
print("Card Transaction Data - Before Scaling:")
print(card_trans_df.describe())
print("\n" + "="*50 + "\n")
print("Credit Card Data - Before Scaling:")
print(creditcard_df.describe())
```

## 2. Apply Scaling Techniques

```python
# Function to scale numeric features
def scale_features(df, scaler_type='standard', exclude_cols=None):
    """
    Scale numeric features in the dataframe
    
    Parameters:
    - df: Input dataframe
    - scaler_type: 'standard', 'minmax', or 'robust'
    - exclude_cols: Columns to exclude from scaling (e.g., target column)
    """
    df_copy = df.copy()
    
    if exclude_cols is None:
        exclude_cols = []
    
    # Select numeric columns
    numeric_cols = df_copy.select_dtypes(include=[np.number]).columns
    numeric_cols = [col for col in numeric_cols if col not in exclude_cols]
    
    # Choose scaler
    if scaler_type == 'standard':
        scaler = StandardScaler()
    elif scaler_type == 'minmax':
        scaler = MinMaxScaler()
    elif scaler_type == 'robust':
        scaler = RobustScaler()
    else:
        raise ValueError("Invalid scaler type")
    
    # Scale the features
    df_copy[numeric_cols] = scaler.fit_transform(df_copy[numeric_cols])
    
    return df_copy, scaler

# Apply Standard Scaler (recommended for most cases)
card_trans_scaled, card_scaler = scale_features(card_trans_df, 'standard')
creditcard_scaled, cc_scaler = scale_features(creditcard_df, 'standard')

print("Features scaled successfully using StandardScaler")
```

```python
# Check data statistics after scaling
print("Card Transaction Data - After Scaling:")
print(card_trans_scaled.describe())
print("\n" + "="*50 + "\n")
print("Credit Card Data - After Scaling:")
print(creditcard_scaled.describe())
```

## 3. Compare Different Scaling Methods

```python
# Compare scaling methods
card_trans_minmax, _ = scale_features(card_trans_df, 'minmax')
card_trans_robust, _ = scale_features(card_trans_df, 'robust')

print("Comparison of scaling methods completed")
print("\nStandard Scaler (mean=0, std=1) - Good for normally distributed data")
print("MinMax Scaler (range 0-1) - Good for bounded data")
print("Robust Scaler - Good for data with outliers")
```
